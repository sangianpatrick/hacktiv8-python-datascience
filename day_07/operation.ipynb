```python
ser_mean = pd.Series(df[i].mean() for i in range(0,100))
ser_min = pd.Series(df[i].min() for i in range(0,100))
ser_max = pd.Series(df[i].max() for i in range(0,100))
```

```python
new_df = pd.DataFrame({'rata-rata': ser_mean, 'min': ser_min, 'max': ser_max})
new_df
```

## Buka Data

```python
import pandas as pd

# buka data trian.csv
data_train = pd.read_csv('train.csv', encoding='latin1', header=None)
data_trainLabels = pd.read_csv('trainLabels.csv', encoding='latin1', header=None)

data_gabungan = pd.concat([data_train, data_trainLabels], axis=1)
data_gabungan
# show read data
# data_train[::-1].head()
```

## Visualisasi

```python
from sklearn.decomposition import PCA

X,y = data_train, data_trainLabels

pca = PCA(n_components=2)
pca.fit(X)
X_reduced = pca.transform(X)
X_reduced.shape
y = y[0].tolist()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='RdYlBu')
```

## Train Model Skenario 1

```python
from sklearn.model_selection import train_test_split
import numpy as np
X,y = data_train, data_trainLabels
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#using svc
from sklearn.svm import SVC
model = SVC()

#train
model.fit(X,y)
```

```python
y_pred = model.predict(X_test)

print(np.all(y_test[0] == y_pred))
```

```python
#check accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test[0], y_pred)
```

```python
#using gradient boosting classifier

from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=20, learning_rate=1, max_features=20, max_depth = 2, random_state = 1)
gb.fit(X, y)

y_pred = gb.predict(X_test)
accuracy_score_test = gb.score(X_test, y_pred)

# print("Learning rates: ", 1)
# print("Accuracy training: ", accuracy_score_training)
print("Accuracy score: ", accuracy_score_test)
```

## Hitung Capres

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt

data = pd.read_csv('Capres2014-2.0.csv', encoding='latin1')
```

```python
#hitung jumlah capres
# data.groupby('Kandidat').Kandidat.nunique().count
capres_count = data[data.Kandidat!='None'].Kandidat.value_counts().count()
capres = data[data.Kandidat!='None'].Kandidat.unique()
capres_count
print("Jumlah Capres: {}".format(capres_count))
# capres
```

## Diagram batang (dengan sentimen)

```python
fig = plt.figure(figsize=(10,50))
fig_dims = (capres_count,2)

row = 0
for name in capres:
    plt.subplot2grid(fig_dims, (row//2,row%2))
    data[data.Kandidat==name].Sentimen.value_counts().plot(kind='bar', title = name)

    row += 1
```

## Vectorizer

```python
from sklearn.model_selection import train_test_split

stopwords = pd.read_csv('stopwords_id.csv')
stopwords = stopwords['#stopwords'].values

X = data.Isi_Tweet
y = data.Sentimen

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
vect = CountVectorizer()
X_train_vec = vect.fit(X_train)
X_train_vec.get_feature_names()
X_train_vec = vect.transform(X_train)
X_train_vec
```

```python
X_test_vec = vect.transform(X_test)
X_test_vec
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect2 = TfidfVectorizer(stop_words=list(stopwords))
X_train_vect2 = vect2.fit_transform(X_train)
X_train_vect2
```

```python
X_test_vect2 = vect2.transform(X_test)
X_test_vect2
```

## Sentimen anlisis  dengan algoritma (SVM, NaiveBayes, RandomForestClasifier, GradientBoosting)

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#svm no stopwords
clf = SVC(kernel='linear')
clf.fit(X_train_vec, y_train)
```

```python
y_pred = clf.predict(X_test_vec)
accuracy_score(y_pred, y_test)
```

```python
#svm with stopwords
clf = SVC(kernel='linear')
clf.fit(X_train_vect2, y_train)
```

```python
y_pred = clf.predict(X_test_vect2)
accuracy_score(y_pred, y_test)
```

```python
#naive bayes no stopword
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
%time nb.fit(X_train_vec, y_train)
```

```python
y_pred = nb.predict(X_test_vec)
accuracy_score(y_pred, y_test)
```

```python
#naive bayes with stopword
%time nb.fit(X_train_vect2, y_train)
```

```python
y_pred = nb.predict(X_test_vect2)
accuracy_score(y_pred, y_test)
```

```python
#gradient boosting no stopword
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=20, learning_rate=1, max_features=20, max_depth = 2, random_state = 1)

gb.fit(X_train_vec, y_train)
```

```python
y_pred = gb.predict(X_test_vec)
accuracy_score(y_pred, y_test)
```

```python
#gradient boosting with stopword

gb = GradientBoostingClassifier(n_estimators=20, learning_rate=1, max_features=20, max_depth = 2, random_state = 1)

gb.fit(X_train_vect2, y_train)
```

```python
y_pred = gb.predict(X_test_vect2)
accuracy_score(y_pred, y_test)
```

```python
#RandomForest no stopword
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=20, max_depth = 2, random_state = 1)
rf.fit(X_train_vec, y_train)
```

```python
y_pred = rf.predict(X_test_vec)
accuracy_score(y_pred, y_test)
```

```python
#RandomForest with stopword
rf = RandomForestClassifier(n_estimators=20, max_depth = 2, random_state = 1)
rf.fit(X_train_vect2, y_train)
```

```python
y_pred = rf.predict(X_test_vect2)
accuracy_score(y_pred, y_test)
```
